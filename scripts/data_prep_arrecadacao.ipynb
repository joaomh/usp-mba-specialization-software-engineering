## Tabela Fato de Arrecadação - Fonte: RFB


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# 1. Criação da Tabela com TRY_CAST e Limpeza de Espaços (TRIM)
spark.sql(f"""
CREATE OR REPLACE TABLE fact_arrecadacao_uf
USING DELTA
PARTITIONED BY (nu_ano, nu_mes)
COMMENT 'Tabela Fato de Arrecadação - Fonte: RFB (Tratada com TRY_CAST para evitar erros)'
AS
SELECT
    CAST(_c0 AS INT) AS nu_ano,
    CAST(_c1 AS STRING) AS nu_mes,
    CAST(_c2 AS STRING) AS sg_uf,
    
    -- Aplicamos TRIM para tirar espaços, os dois REPLACE para limpar a pontuação 
    -- e o TRY_CAST para não quebrar caso venha um texto onde deveria ser número
    TRY_CAST(REPLACE(REPLACE(TRIM(_c3), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_imposto_importacao,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c4), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_imposto_exportacao,
    
    -- IPI
    TRY_CAST(REPLACE(REPLACE(TRIM(_c5), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipi_fumo,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c6), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipi_bebidas,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c7), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipi_automoveis,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c8), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipi_vinculado_importacao,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c9), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipi_outros,
    
    -- Imposto de Renda
    TRY_CAST(REPLACE(REPLACE(TRIM(_c10), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irpf,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c11), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irpj_entidades_financeiras,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c12), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irpj_demais_empresas,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c13), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irrf_rendimentos_trabalho,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c14), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irrf_rendimentos_capital,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c15), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irrf_remessas_exterior,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c16), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_irrf_outros_rendimentos,
    
    -- Outros Tributos
    TRY_CAST(REPLACE(REPLACE(TRIM(_c17), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_iof,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c18), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_itr,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c19), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_ipmf,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c20), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cpmf,
    
    -- Contribuições Sociais (COFINS/PIS/CSLL)
    TRY_CAST(REPLACE(REPLACE(TRIM(_c21), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cofins_total,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c22), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cofins_financeiras,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c23), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cofins_demais,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c24), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_pis_pasep_total,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c25), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_pis_pasep_financeiras,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c26), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_pis_pasep_demais,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c27), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_csll_total,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c28), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_csll_financeiras,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c29), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_csll_demais,
    
    -- CIDE e Previdência
    TRY_CAST(REPLACE(REPLACE(TRIM(_c30), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cide_combustiveis_nao_dedutivel,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c31), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cide_combustiveis,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c32), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cpsss_antigo,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c33), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_cpsss_atual,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c34), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_fundaf,
    
    -- Parcelamentos e Outros
    TRY_CAST(REPLACE(REPLACE(TRIM(_c35), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_refis,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c36), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_paes,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c37), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_retencao_fonte_lei_10833,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c38), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_pagamento_unificado,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c39), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_outras_receitas_administradas,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c40), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_demais_receitas,
    
    -- Receita Previdenciária
    TRY_CAST(REPLACE(REPLACE(TRIM(_c41), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_receita_previdenciaria_total,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c42), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_receita_previdenciaria_propria,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c43), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_receita_previdenciaria_demais,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c44), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_receitas_outros_orgaos

FROM read_files(
    '/Volumes/workspace/default/raw/fact_arrecadacao_estado/*.csv',
    format => 'csv',
    delimiter => ';',
    header => 'false',
    encoding => 'ISO-8859-1'
)
""")
```

## Criação da Tabela Fato Arrecadação IR e IPI

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# 1. Criação da Tabela Fato Arrecadação IR e IPI
spark.sql(f"""
CREATE OR REPLACE TABLE fact_arrecadacao_ir_ipi
USING DELTA
PARTITIONED BY (nu_ano, nu_mes)
COMMENT 'Dados de arrecadação decendial de IR e IPI - Fonte: RFB DW Arrecadação'
AS
SELECT    
    -- Identificação e Período
    CAST(TRIM(_c0) AS STRING) AS ds_tributo,
    CAST(_c1 AS INT) AS nu_ano,
    CAST(_c2 AS STRING) AS nu_mes,
    CAST(_c3 AS STRING) AS nu_decendio,
    
    -- Valores Monetários (Limpando ponto de milhar e vírgula decimal)
    TRY_CAST(REPLACE(REPLACE(TRIM(_c4), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_arrecadacao_bruta,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c5), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_retificacao,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c6), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_compensacao,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c7), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_restituicao,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c8), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_outros,
    TRY_CAST(REPLACE(REPLACE(TRIM(_c9), '.', ''), ',', '.') AS DECIMAL(18,2)) AS vl_arrecadacao_liquida

FROM read_files(
    '/Volumes/workspace/default/raw/fact_arrecadacao_ir_ipi/*.csv',
    format => 'csv',
    delimiter => ';',
    header => 'false',
    encoding => 'ISO-8859-1'
)
""")

# 2. Documentação das colunas baseada nos metadados
column_comments = {
    "ds_tributo": "Tributo, IR ou IPI",
    "nu_ano": "Ano de referência ",
    "nu_mes": "Mês de referência ",
    "nu_decendio": "Decêndio de referência (1º, 2º ou 3º período de 10 dias) ",
    "vl_arrecadacao_bruta": "Valor da arrecadação bruta sem acréscimos ou deduções ",
    "vl_retificacao": "Valor do saldo de retificações efetuadas nos Darf ",
    "vl_compensacao": "Valor do saldo das compensações efetuadas ",
    "vl_restituicao": "Valor das restituições por pagamentos indevidos ou a maior ",
    "vl_outros": "Demais eventos que agem sobre a arrecadação bruta (ex: cancelamentos)",
    "vl_arrecadacao_liquida": "Valor informado à STN após incidência de todos os eventos"
}

for col_name, comment in column_comments.items():
    spark.sql(f"ALTER TABLE fact_arrecadacao_ir_ipi ALTER COLUMN {col_name} COMMENT '{comment}'")

print("Tabela fact_arrecadacao_ir_ipi criada com sucesso!")
```
